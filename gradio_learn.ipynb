```python
import gradio as gr

def greet(name, cnt):
    return "Hello {name}".format(name=name) * cnt

demo = gr.Interface(
    fn=greet,
    inputs=["text", "slider"],
    outputs=["text"],
    api_name="predict"
)

demo.launch()
```

```python
import gradio as gr

def greet(name):
    return f"Hello {name}"

with gr.Blocks() as demo:
    name = gr.Textbox(label="name")
    output = gr.Textbox(label="Output Box")
    greet_btn = gr.Button("Greet")
    greet_btn.click(fn=greet, inputs=name, outputs=output, api_name="greet")

demo.launch()
```
