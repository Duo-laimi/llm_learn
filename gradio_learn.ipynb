```python
import gradio as gr

def greet(name):
    return f"Hello {name}"

with gr.Blocks() as demo:
    name = gr.Textbox(label="name")
    output = gr.Textbox(label="Output Box")
    greet_btn = gr.Button("Greet")
    greet_btn.click(fn=greet, inputs=name, outputs=output, api_name="greet")

demo.launch()
```

```python
import gradio as gr
with gr.Blocks() as demo:
    name = gr.Textbox(label="name")
    output = gr.Textbox(label="Output Box", interactive=True)
    greet_btn = gr.Button("Greet")

    @greet_btn.click(inputs=name, outputs=output, api_name="greet")
    def greet(name):
        return f"Hello {name}!"

demo.launch()
```

```python
import gradio as gr

def greet(name):
    return f"hello {name}"


with gr.Blocks() as demo:
    gr.Markdown(
        """
        # Hello World!
        Start typing below to see the output.
        """
    )
    inputs = gr.Textbox(placeholder="please input your name.")
    outputs = gr.Textbox()

    inputs.change(greet, inputs, outputs)

demo.launch()

```

```python
import gradio as gr

def increase(num):
    return num + 1

with gr.Blocks() as demo:
    a = gr.Number(label="input a")
    b = gr.Number(label="input b")
    a_btn = gr.Button("a > b")
    b_btn = gr.Button("b > a")
    a_btn.click(increase, a, b)
    b_btn.click(increase, b, a)

demo.launch()
```

```python
import os

# 必须在导入 transformers 之前设置
os.environ['HF_HOME'] = '/root/lanyun-tmp/model'
os.environ["HF_ENDPOINT"] = "https://hf-mirror.com"

from transformers import pipeline

import gradio as gr

asr = pipeline("automatic-speech-recognition", "facebook/wav2vec2-base-960h")
classifier = pipeline("text-classification")

def speech_to_text(speech):
    text = asr(speech)["text"]
    return text

def text_to_sentiment(text):
    return classifier(text)[0]["label"]

demo = gr.Blocks()

with demo:
    audio_file = gr.Audio(type="filepath")
    text = gr.Textbox()
    label = gr.Label()

    b1 = gr.Button("Recognize Speech")
    b2 = gr.Button("Classify Sentiment")

    b1.click(speech_to_text, inputs=audio_file, outputs=text)
    b2.click(text_to_sentiment, inputs=text, outputs=label)

demo.launch()
```
